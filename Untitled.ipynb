```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Update paths as needed if not running inside WSL
csv_paths = [
    "/tmp/ML-8B-DO-FS-test.csv",
    "/tmp/ML-8B-BC-FS-test.csv",
    "/tmp/ML-8B-HC-FS-test.csv"
]


labels = ['DO', 'BC', 'HC']

dfs = [pd.read_csv(path) for path in csv_paths]
for df, label in zip(dfs, labels):
    df['run_label'] = label  # for tracking in plots
```

```python
df_all = pd.concat(dfs, ignore_index=True)
```

```python
print(df_all.groupby('run_label')['is_perfect_match'].mean())
# No perfects - Some code that I randomly selected was correct, but others had code that was not 
# ( code did work for a vast majority though )

print(df_all.groupby('run_label')['time_taken_seconds'].mean())
# Average time taken

print(df_all.groupby('run_label')[['summary_tokens', 'overall_tokens']].mean())
# Token counts, this is NOT as predicted. I expected DO ( Diverse Output ), to have the most overall tokens
# I predicted HC ( High Consistency ) to be the lowest token wise.
# I did not think the summary tokens would be nearly identical. 
# Summary_tokens vs Overall_tokens = 
# Summary tokens are - Tokens from summary
# Overall tokens are - Summary tokens + Tokens from the Code response ( Overall Tokens are being recorded here )
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_all, x='run_label', y='time_taken_seconds')
plt.title('Time Taken per Run')
plt.ylabel('Time (seconds)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_all, x='run_label', y='summary_tokens')
plt.title('Summary Tokens per Run')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df_all, x='run_label', y='overall_tokens')
plt.title('Overall Tokens per Run')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_all, x='summary_tokens', y='time_taken_seconds', hue='run_label')
plt.title('Time Taken vs. Summary Tokens')
plt.show()
```

```python
pivot_pm = df_all.pivot(index='query_id', columns='run_label', values='overall_tokens')
pivot_time = df_all.pivot(index='query_id', columns='run_label', values='time_taken_seconds')

# Differences
pivot_pm['delta_DO_vs_BC'] = pivot_pm['DO'] - pivot_pm['BC']
pivot_pm['delta_DO_vs_HC'] = pivot_pm['DO'] - pivot_pm['HC']

print(pivot_pm[['delta_DO_vs_BC', 'delta_DO_vs_HC']].describe())

# Plot differences
pivot_pm[['delta_DO_vs_BC', 'delta_DO_vs_HC']].plot(kind='hist', bins=20, alpha=0.7)
plt.title('Distribution of Perfect Match Differences')
plt.xlabel('Difference')
plt.show()
```
